```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

norm = pd.Series([np.random.randn() for i in range(1, 200)])

fig = plt.figure(figsize=(32, 32))
spec = gridspec.GridSpec(ncols=4, nrows=2, figure=fig)
ax1 = fig.add_subplot(spec[0, 0])
ax2 = fig.add_subplot(spec[0, 1])
ax3 = fig.add_subplot(spec[1, 0])
ax4 = fig.add_subplot(spec[1, 1])

norm.plot(kind='hist', ax=ax1)
```

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

norm = pd.Series([np.random.randn() for i in range(1, 200)])
small = pd.Series([abs(np.random.randn()) for i in range(1, 10)])
r = pd.Series(np.random.rand(200))
```

```python
fig, axes = plt.subplots(ncols=4, nrows=3, figsize=(18.6, 14))

norm.sort_values(ignore_index=True).plot(kind='line', ax=fig.axes[0])
fig.axes[0].set_title('Line (line)')

small.plot(kind='bar', ax=fig.axes[1])
fig.axes[1].set_title('Vertical Bars (bar)')

small.plot(kind='barh', ax=fig.axes[2])
fig.axes[1].set_title('Horizontal Bars (barh)')

norm.plot(kind='hist', ax=fig.axes[3])
fig.axes[3].set_title('Histogram (hist)')

norm.plot(kind='box', ax=fig.axes[4])
fig.axes[4].set_title('Boxplot (box)')

norm.plot(kind='kde', ax=fig.axes[5])
fig.axes[5].set_title('Kernel Density Estimation (kde)')

small.plot(kind='area', ax=fig.axes[6])
fig.axes[6].set_title('Area under line (area)')

small.plot(kind='pie', ax=fig.axes[7])
fig.axes[7].set_title('Pie (pie)')

pd.DataFrame({'norm':norm, 'r':r}).plot(kind='scatter', ax=fig.axes[8], x='r', y='norm')
fig.axes[8].set_title('Scatter plot (scatter)')

pd.DataFrame({'norm':norm, 'r':r}).plot(kind='hexbin', ax=fig.axes[9], x='r', y='norm', gridsize = 16)
fig.axes[9].set_title('Hexbin (hexbin)')

fig.axes[10].set_axis_off()
fig.axes[11].set_axis_off()

fig.savefig('kinds_of_pyplot.jpg')
```
